# Predator-Prey by explicit time integrators

```python
import numpy as np
import matplotlib.pyplot as plt
from explicit_methods import explicit_euler
from explicit_methods import modified_euler
from explicit_methods import heun
```

```python
b0=0.75
r0=0.33
T=30
N=1000
lambda_b=1
b_e=0.5
lambda_r=1.5
r_e=0.25
dt=T/N
```

```python
def f(t,u):
    f1=lambda_b*u[0]*(1-u[1]/r_e)
    f2=lambda_r*u[1]*(u[0]/b_e-1)
    f=np.array([f1, f2])
    return f
```

```python
# initial setup
t = np.linspace(0, T, N+1)
u_euler = np.zeros((2, N+1))
u_mod=np.zeros((2, N+1))
u_heun=np.zeros((2, N+1))

# initial value
u_euler[:,0]=np.array([b0, r0])
u_mod[:,0]=np.array([b0, r0])
u_heun[:,0]=np.array([b0, r0])
```

```python
for step in range(N):
    u_euler[:,step+1]=explicit_euler(u_euler[:,step], t[step], dt, f)
    u_mod[:,step+1]=modified_euler(u_mod[:,step], t[step], dt, f)
    u_heun[:,step+1]=heun(u_heun[:,step], t[step], dt, f)
    
```

```python
# Plot the populations over time
plt.figure()
plt.plot(t, u_euler[0,:], 'g', label='Euler Prey')
plt.plot(t, u_mod[0,:], 'g--', linewidth=2, label='Modified Euler Prey')
plt.plot(t, u_heun[0,:], 'g:', linewidth=5, label='Heun Prey') 
plt.plot(t, u_euler[1,:], 'r', label='Euler Predator')
plt.plot(t, u_mod[1,:], 'r--', label='Modified Euler Predator')
plt.plot(t, u_heun[1,:], 'r:', linewidth=5, label='Heun Predator')
plt.xlabel('Time')
plt.ylabel('Population')
plt.legend()
plt.title('Predator and Prey Populations')
plt.grid(True)

# Plotting in phase space
plt.figure()
plt.plot(u_euler[0,:], u_euler[1,:], 'b', label='Euler')
plt.plot(u_mod[0,:], u_mod[1,:], 'b--', linewidth=2, label='Modified Euler')
plt.plot(u_heun[0,:], u_heun[1,:], 'b:', linewidth=5, label='Heun')
plt.xlabel('Prey Population')
plt.ylabel('Predator Population')
plt.legend()
plt.title('Phase Space')
plt.grid(True)
plt.show()
```
